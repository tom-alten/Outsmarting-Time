## Config

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Data

### Results from eval scripts

```python
path_to_year = '/Users/tomaltenborg/Documents/Master/Master thesis/Notebooks/M-Competitions/M3-Competition/Year/M3_yearly_results.parquet'
path_to_quarter = '/Users/tomaltenborg/Documents/Master/Master thesis/Notebooks/M-Competitions/M3-Competition/Quarter/M3_quarter_results.parquet'
path_to_month = '/Users/tomaltenborg/Documents/Master/Master thesis/Notebooks/M-Competitions/M3-Competition/Month/M3_monthly_results.parquet'

df_year = pd.read_parquet(path_to_year)
df_quarter = pd.read_parquet(path_to_quarter)
df_month = pd.read_parquet(path_to_month)
```

```python
df = pd.concat([df_year, df_quarter, df_month])
df = df.reset_index(drop=True)
df.sort_values(by = ['Series', 'ForecastPeriod', 'Method'], inplace = True)
```

## Visualizations all data

```python
methods = ['ARARMA', 'Auto-ANN', 'AutoBox1', 'B-J auto',
 'COMB S-H-D', 'DAMPEN', 'Flors-Pearc1', 'ForcX', 'ForecastPro',
 'HOLT', 'NAIVE2', 'PP-Autocast', 'RBF', 'ROBUST-Trend', 'SINGLE', 'SMARTFCS',
 'THETA', 'TimeGPT', 'WINTER', 'AAM1']
```

### Average sMAPE for all data

```python
filtered_df = df[df['Method'].isin(methods)]

# Calculate mean of sMAPE per unique Method
mean_sMAPE_per_method = filtered_df.groupby('Method')['sMAPE'].mean().sort_values()

# Plot
plt.figure(figsize=(12, 8))  # Set figure size
bars = mean_sMAPE_per_method.plot(kind='bar', color=['blue' if method != 'TimeGPT' else 'red' for method in mean_sMAPE_per_method.index])
plt.title('Mean sMAPE per All Data (year, quarter and month)', fontsize = 16, fontweight = 'bold')  # Title of the chart
plt.ylabel('Mean sMAPE', fontsize = 12)  # Y-axis label
plt.xlabel('Method', fontsize = 12)  # X-axis label
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels and align them to the right for better readability
plt.tight_layout()  # Adjust layout to make sure everything fits without overlapping

# Customize colors
for method, bar in zip(mean_sMAPE_per_method.index, bars.patches):
    bar.set_color('red' if method == 'TimeGPT' else 'blue')

plt.show()
```

### Average RMSE for all data 

```python
filtered_df = df[df['Method'].isin(methods)]

# Calculate mean of RMSE per unique Method
mean_RMSE_per_method = filtered_df.groupby('Method')['RMSE'].mean().sort_values()

# Plot
plt.figure(figsize=(12, 8))  # Set figure size
bars = mean_RMSE_per_method.plot(kind='bar', color=['blue' if method != 'TimeGPT' else 'red' for method in mean_RMSE_per_method.index])
plt.title('Mean RMSE All Data (year, quarter and month)', fontsize = 16, fontweight = 'bold')  # Title of the chart
plt.ylabel('Mean RMSE', fontsize = 12)  # Y-axis label
plt.xlabel('Method', fontsize = 12)  # X-axis label
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels and align them to the right for better readability
plt.tight_layout()  # Adjust layout to make sure everything fits without overlapping

# Customize colors
for method, bar in zip(mean_RMSE_per_method.index, bars.patches):
    bar.set_color('red' if method == 'TimeGPT' else 'blue')

plt.show()
```
