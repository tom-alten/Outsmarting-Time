```python
def compute_rmsse(Y_actual, Y_forecast, Y_train):
    """
    Compute the RMSSE for a single time series.

    Parameters:
    - Y_actual: Actual values for the forecast horizon (numpy array or list).
    - Y_forecast: Forecasted values for the forecast horizon (numpy array or list).
    - Y_train: Historical observations used for training (numpy array or list).

    Returns:
    - rmsse: The RMSSE score.
    """
    # Ensure inputs are numpy arrays for vectorized operations
    Y_actual = np.array(Y_actual)
    Y_forecast = np.array(Y_forecast)
    Y_train = np.array(Y_train)
    
    # Calculate the numerator: sum of squared forecast errors
    forecast_errors = Y_actual - Y_forecast
    squared_forecast_errors = forecast_errors ** 2
    numerator = np.sum(squared_forecast_errors)
    
    # Calculate the denominator: mean of squared differences in the training data
    # Consider only periods following the first non-zero demand
    Y_train_nonzero = Y_train[np.argmax(Y_train != 0):]  # Exclude leading zeros
    squared_diffs = np.diff(Y_train_nonzero) ** 2
    denominator = np.mean(squared_diffs)
    
    h = len(Y_actual)  # Forecast horizon
    n = len(Y_train_nonzero)  # Effective length of training sample
    
    # Compute RMSSE
    rmsse = np.sqrt((1/h) * (numerator / ((1/(n-1)) * denominator)))
    
    return rmsse
```
